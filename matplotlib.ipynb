# <font color=hotpink >matplotlib</font>

- python library for data visualization
- create 2D graphs and plots by using python scripts
- produces output in a variety of hardcopy formats

## Features
- Provides a module called pyplot
- Simple functions used for visualization
- Supports a very wide variety of graphs
- Easy integration with Pandas and Numpy
- Provides an Object-Oriented API

```python

import numpy as np
from matplotlib import pyplot as plt
```

## line plot
- relationship between two numerical entities

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

### adding title and labels

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Changing Line Aesthetics

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

### Adding two lines in the same plot

```python
x
```

```python
y1= 2*x
y2=3*x
```

```python
plt.plot(x,y1,color="g",linewidth=2)
plt.plot(x,y2,color="r",linewidth=5)
plt.grid(True)
plt.show()
```

### Adding sub-plots

```python
x=np.arange(1,11)
y1= 2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color="yellow")
plt.subplot(1,2,2)
plt.plot(x,y2,color="orange")
plt.show()

```

```python
# plot along the rows

plt.subplot(2,1,1)
plt.plot(x,y1,color="yellow")
plt.subplot(2,1,2)
plt.plot(x,y2,color="orange")
plt.show()


```

# Bar plot

- helps us to understand the disturbution of categorical column

```python
student ={"bob":87,"matt":97,"sam":23}

names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
        
```

### adding title and labels

```python
plt.bar(names,values)
plt.title("Marks of Students")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

## Horizantal Bar Plot

```python
plt.barh(names,values,color='g')
plt.title("Marks of Students")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.grid(True)
plt.show()
```

# scatter plot

-  used to understand the distribution between two numerical entities and these entities are represented in the form ofdata points

```python
x =[ 1,2,3,4,5,6,7,8,9]
y= [5,26,8,1,3,4,1,8,7]
# x and y values must be same

plt.scatter(x,y,color="maroon")
plt.show()
```

## Changing Mark Aesthetics

```python
x =[ 1,2,3,4,5,6,7,8,9]
y= [5,26,8,1,3,4,1,8,7]

plt.scatter(x,y,marker="*",color="violet",s=700)
plt.show()
```

## Adding two markers in the same plot

```python
x =[ 10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]

plt.scatter(x,a,marker="*",color="pink",s=200)

plt.scatter(x,b,marker="x",color="yellow",s=100)
plt.show()
```

## Adding sub-plots

```python
x =[ 10,20,30,40,50,60,70,80,90]
a= [8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",color="brown",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",color="yellow",s=200)
plt.show()
```

# pie chart
- helps us to understand the frequencies or percentage of different categorical values

```python
# creating data
fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]
```

```python
#making plot
plt.pie(quantity,labels=fruit)
plt.show()
```

## Changing Aesthetics

```python

fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit,autopct='%0.1f%%' ,colors=["red","orange","yellow","green"])
plt.show()
```

# DoughNut-Chart

```python
# creating data
fruit=['Apple','Orange','Mango','Guava','lichi']
quantity=[67,34,100,29,112]
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%' ,colors=["red","orange","yellow","green","pink"],radius=2)
plt.pie([5],colors=['w'],radius=1)
plt.show()
```
